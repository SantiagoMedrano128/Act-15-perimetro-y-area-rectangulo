### Redaccion del problema
Se solicita desarrollar un programa interactivo que permita al usuario realizar cálculos básicos sobre un rectángulo. El sistema debe mostrar un menú persistente que permita calcular el área o el perímetro, utilizando funciones específicas para cada operación y validando la entrada del usuario mediante estructuras de control.

#### Formulas
1. Área: $A = b \times h$
2. Perímetro: $P = 2 \times (b + h)$

![texto vinculado](https://raw.githubusercontent.com/SantiagoMedrano128/Act-15-perimetro-y-area-rectangulo/refs/heads/main/El%20rectangulo.webp)

```python
def calcular_area(base, altura):
    """Calcula y retorna el área del rectángulo."""
    return base * altura

def calcular_perimetro(base, altura):
    """Calcula y retorna el perímetro del rectángulo."""
    return 2 * (base + altura)

def menu():
    """Maneja la lógica de la interfaz y llamadas a funciones."""
    continuar = True

    while continuar:
        print("\n--- 📏 MENÚ RECTÁNGULO 📐 ---")
        print("1. 🟦 Calcular área")
        print("2. 🔄 Calcular perímetro")
        print("3. ❌ Salir")

        opcion = input("\n➤ Seleccione una opción: ")

        if opcion == '1' or opcion == '2':
            # Solicitamos datos al usuario
            try:
                b = float(input("⌨️  Ingrese la base: "))
                h = float(input("⌨️  Ingrese la altura: "))

                if opcion == '1':
                    resultado = calcular_area(b, h)
                    print(f"✅ El área es: {resultado} u² 🟦")
                else:
                    resultado = calcular_perimetro(b, h)
                    print(f"✅ El perímetro es: {resultado} u 🔄")
            except ValueError:
                print("⚠️ Error: Por favor, ingrese solo números.")

        elif opcion == '3':
            print("👋 Saliendo del programa... ¡Hasta pronto! ✨")
            print("Programa realizado por Santiago Medrano NL = 37")
            continuar = False

        else:
            print("🚫 Opción no válida. Intente de nuevo. 🧐")

# Iniciar el programa
if __name__ == "__main__":
    menu()
```
